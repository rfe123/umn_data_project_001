```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
from pprint import pprint
```

```python
sleep_health_df = pd.read_csv('../Resources/Sleep_health_and_lifestyle_dataset.csv')

sleep_health_df
```

```python
x_axis = 'Sleep Duration'
y_axis = 'Daily Steps'

sleep_health_df.plot.scatter(x_axis, y_axis)
plt.title(f"{x_axis} vs. {y_axis}")
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.show()
```

```python
x_axis = 'Sleep Duration'
y_axis = 'Quality of Sleep'

sleep_health_df.plot.scatter(x_axis, y_axis)
plt.title(f"{x_axis} vs. {y_axis}")
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.show()
```

```python
x_axis = 'Sleep Duration'
y_axis = 'Quality of Sleep'

male_df = sleep_health_df.loc[(sleep_health_df['Gender']=='Male')]
female_df = sleep_health_df.loc[(sleep_health_df['Gender']=='Female')]

plot1 = male_df.plot.scatter(x_axis, y_axis)
female_df.plot.scatter(x_axis, y_axis, color='red', ax=plot1)
plt.title(f"{x_axis} vs. {y_axis}")
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.show()
```

```python
x_axis = 'Stress Level'
y_axis = 'Quality of Sleep'

male_df = sleep_health_df.loc[(sleep_health_df['Gender']=='Male')]
female_df = sleep_health_df.loc[(sleep_health_df['Gender']=='Female')]

plot1 = male_df.plot.scatter(x_axis, y_axis)
female_df.plot.scatter(x_axis, y_axis, color='red', ax=plot1)
plt.title(f"{x_axis} vs. {y_axis}")
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.show()
```

```python
x_axis = 'Sleep Duration'
y_axis = 'Daily Steps'

plot2 = male_df.plot.scatter(x_axis, y_axis)
female_df.plot.scatter(x_axis, y_axis, color='red', ax=plot2)
plt.title(f"{x_axis} vs. {y_axis}")
plt.xlabel(x_axis)
plt.ylabel(y_axis)

plt.show()
```

```python
y_axis = 'Stress Level'
x_axis = 'Blood Pressure'

sleep_health_df.sort_values('Blood Pressure').plot.scatter(x_axis, y_axis)

plt.title(f"{x_axis} vs. {y_axis}")
plt.xlabel(x_axis)
plt.ylabel(y_axis)
plt.xticks(rotation=45)
plt.show()
```

```python
sleep_health_df[['Systolic','Diastolic']] = sleep_health_df['Blood Pressure'].str.split('/', expand=True)
sleep_health_df.head()
```

```python
y_axis = 'Stress Level'
x_axis = 'Systolic'

sleep_health_df.sort_values('Blood Pressure').plot.scatter(x_axis, y_axis)

plt.title(f"{x_axis} vs. {y_axis}")
plt.xlabel(x_axis)
plt.ylabel(y_axis)
plt.xticks(rotation=45)
plt.show()
```
